```python
#loading .env variables

import os
from dotenv import load_dotenv
load_dotenv()
```

```python
#Initializing open ai embeddings
from langchain_openai import OpenAIEmbeddings
embeddings=OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text="This is tutorial on open ai embedding"
query_result=embeddings.embed_query(text)
query_result
```

```python
#we can change the dimensions as we want

#Initializing open ai embeddings
from langchain_openai import OpenAIEmbeddings
embeddings_1024=OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)
embeddings_1024
```

```python
text="This is tutorial on open ai embedding"
query_result_1024=embeddings_1024.embed_query(text)
query_result_1024
```

```python
len(query_result_1024)
```

#### Taking speech.txt as example

```python
# Text loader helps us to load any txt file

from langchain_community.document_loaders import TextLoader

#Initializing text loader

loader = TextLoader("speech.txt")
text_documents = loader.load()
text_documents
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_documents = text_splitter.split_documents(text_documents)
final_documents
```

```python
from langchain_community.vectorstores import Chroma
db = Chroma.from_documents(final_documents,embeddings_1024)
db
```

```python
#How to find the vectors of perticular text..?
#copy some content from speech.txt and 

query = "It will be all the easier for us to conduct ouselves as "
result = db.similarity_search(query)
print(result)
```
